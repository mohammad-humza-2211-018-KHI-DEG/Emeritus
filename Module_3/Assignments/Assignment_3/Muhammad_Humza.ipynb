```python
import matplotlib.pyplot as plt
import numpy as np
from sklearn import datasets
from sklearn.cluster import KMeans
from sklearn.decomposition import PCA
```

```python
iris = datasets.load_iris()
```

# K-Means With-out PCA

```python
x_Kmeans = iris.data
y_Kmeans = iris.target
```

```python
x_Kmeans.shape
```

```python
model1 = KMeans(n_clusters=3, n_init=1, max_iter=100)
model1.fit(x_Kmeans)

all_predictions_Kmeans = model1.predict(x_Kmeans)
centroids_Kmeans = model1.cluster_centers_
```

```python
centroids_Kmeans
```

```python
plt.scatter(x_Kmeans[:,0], x_Kmeans[:,1], c=all_predictions_Kmeans)
plt.scatter(centroids_Kmeans[:,0], centroids_Kmeans[:,1], marker='x', color="black")
plt.show
```

# Applying PCA

```python
pca = PCA(n_components=2)
x_reduced = pca.fit_transform(x_Kmeans)

x_reduced.shape
```

# K-Means With PCA

```python
model = KMeans(n_clusters=3, n_init=1, max_iter=100)
model.fit(x_reduced)

all_predictions_PCA = model.predict(x_reduced)
centroids_PCA = model.cluster_centers_
```

```python
centroids_PCA
```

```python
plt.scatter(x_reduced[:,0], x_reduced[:,1], c=all_predictions_PCA)
plt.scatter(centroids_PCA[:,0], centroids_PCA[:,1], marker='x', color="black")
plt.show
```

# Result Comparison 

***Calculating the mean of all distances from the centroid.***

```python
#K_Means
dist = np.linalg.norm(x_Kmeans - centroids_Kmeans[0,:],axis=1).reshape(-1,1)
dist.mean()
```

```python
#PCA
dist = np.linalg.norm(x_reduced - centroids_PCA[0,:],axis=1).reshape(-1,1)
dist.mean()
```
