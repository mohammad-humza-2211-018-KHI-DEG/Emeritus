```python
import pandas as pd
import numpy as np
```

```python
day_10 = pd.read_parquet('earnings_date=2022-02-10/employee_earnings.parquet', engine='pyarrow')
day_11 = pd.read_parquet('earnings_date=2022-02-11/employee_earnings.parquet', engine='pyarrow')
day_12 = pd.read_parquet('earnings_date=2022-02-12/employee_earnings.parquet', engine='pyarrow')
day_13 = pd.read_parquet('earnings_date=2022-02-13/employee_earnings.parquet', engine='pyarrow')
day_14 = pd.read_parquet('earnings_date=2022-02-14/employee_earnings.parquet', engine='pyarrow')
```

```python
df_f = day_10.copy()
df_f.head()
```

```python
df = pd.DataFrame()
```

```python
df['earnings_10'] = day_10['earnings']
df['earnings_11'] = day_11['earnings']
df['earnings_12'] = day_12['earnings']
df['earnings_13'] = day_13['earnings']
df['earnings_14'] = day_14['earnings']
```

```python
df.head()
```

```python
df1 = df.copy()
df1['mean'] = df1.mean(axis=1)
```

```python
df1['Increment'] = df1.std(axis=1)
```

```python
df1['ratio'] = df1['Increment']/df1['mean']
```

```python
df1['earnings_15'] = df1['Increment'] + df1['earnings_14']
```

```python
df1['earnings_16'] = df1['Increment'] + df1['earnings_15']
```

```python
df1.head()
```

```python
day_15 = day_10.copy()
```

```python
day_15['earnings'] = df1['earnings_15'].astype(int)
```

```python
day_15.head()
```

```python
day_15.to_parquet('employee_earnings.parquet')
```
